```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as  sns

p = lambda: plt.show()
```

### Attention

3 vectors for `key`, `query`, `value`, init with random

```python
# means 5 words with 3 dimensions for vector represent
k = np.random.randn(5, 3)
q = np.random.randn(5, 3)
v = np.random.randn(5, 3)
```

```python
# softmax
def softmax(x):
    e_x = np.exp(x)
    return e_x / e_x.sum(axis=-1)

```

```python
score = np.matmul(q, k.transpose(1, 0)) / np.sqrt(k.shape[-1])

score = softmax(score)

out = np.matmul(score, v)
```

```python
# plot
sns.heatmap(score, annot=True)
# as there are 5 words, the heatmap means for each word will contribute for each word score
p()
```

```python
sns.heatmap(out, annot=True)
# so out is the represent for the input with 5 words with attention 
p()
```
